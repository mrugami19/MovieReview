```python
def load_data():
    print("Loading the imdb reviews data...")
    train_corpus=[]
    y_train=[]
    directory = os.path.normpath("/Users/rumi/desktop/pos")
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".txt"):
                f=open(os.path.join(subdir, file),'r')
                try:
                    a = f.read() 
                    train_corpus.append(a)
                    y_train.append(0)
                except:
                    pass
                f.close()
    directory = os.path.normpath("/Users/rumi/desktop/neg")
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".txt"):
                f=open(os.path.join(subdir, file),'r')
                try:
                    a = f.read() 
                    train_corpus.append(a)
                    y_train.append(1)
                except:
                    pass
                f.close()
    
    print("Data loaded successfully!")
    return train_corpus, y_train
```

```python
train_corpus, y_train=load_data()
```

```python
#separating sentences from the reviews
train_sent=[]
train_review=[]
for review in train_corpus:
    train_review.append(re.split(r'[.!?]', review ))
for rev in train_review:
    for sent in rev:
        train_sent.append(sent)
```
